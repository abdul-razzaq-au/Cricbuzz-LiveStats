```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

base_url = "https://en.wikipedia.org"
list_url = "https://en.wikipedia.org/wiki/List_of_international_cricket_grounds"

response = requests.get(list_url)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table", class_="wikitable")

grounds_data = []

for table in tables:
    rows = table.find_all("tr")[1:]

    for row in rows:
        cols = row.find_all("td")
        if len(cols) < 1:
            continue

        link = cols[0].find("a")
        if link:
            ground_name = link.text.strip()
            ground_link = base_url + link["href"]

            try:
                ground_page = requests.get(ground_link)
                ground_soup = BeautifulSoup(ground_page.text, "html.parser")

                info_box = ground_soup.find("table", class_="infobox")

                capacity = None
                city = None
                country = None

                if info_box:
                    rows_info = info_box.find_all("tr")
                    for r in rows_info:
                        header = r.find("th")
                        value = r.find("td")

                        if header and value:
                            header_text = header.text.strip()

                            if "Capacity" in header_text:
                                capacity_text = value.text.strip()
                                capacity = re.sub(r"\[.*?\]", "", capacity_text)

                            if "Location" in header_text:
                                location_text = value.text.strip()
                                location_text = re.sub(r"\[.*?\]", "", location_text)
                                city = location_text

                grounds_data.append({
                    "Ground": ground_name,
                    "Location": city,
                    "Capacity": capacity
                })

                print(f"Scraped: {ground_name}")
                time.sleep(1)

            except:
                print(f"Failed: {ground_name}")
                continue

df = pd.DataFrame(grounds_data)
print(df.head())

df.to_csv("cricket_grounds.csv", index=False)
```

```python
win_count = defaultdict(int)  # dictionary to store win count
for typematch in data5.get('typeMatches', []):
    for seriesmatch in typematch.get('seriesMatches', []):
        series_info = seriesmatch.get('seriesAdWrapper', {})
        for match in series_info.get('matches', []):
            matchinfo = match.get('matchInfo', {})
            status = matchinfo.get('status', '')

            if 'won by' in status:
                winner = status.split('won')[0]
                win_count[winner] += 1

# sorting

sorted_win = sorted(win_count.items(), key=lambda x:x[1], reverse=True)

# display win result

for i,j in win_count.items():
    print(f"{i} - {j}")
```
